```python
from sktime.transformations.panel.rocket import MiniRocketMultivariate
from sklearn.linear_model import RidgeClassifierCV
```

```python
import numpy as np
from context import Context
from data import Data
from copy import deepcopy
from baselines import cross_val
from util import Metric

Context.data_dir = "/home/foumani/data/solar_flare_prediction"
Context.files_df_filename = "all_files.csv"
Context.files_np_filename = "full_data_X_1_25.npy"
context = Context(train_n=[1500, 800], val_part=1, test_part=2)
data = Data()
```

```python
def baseline_minirocket(X_train, y_train, X_val, y_val, X_test, y_test):
    minirocket = MiniRocketMultivariate()
    minirocket.fit(X_train)
    X_train_transform = minirocket.transform(X_train)
    classifier = RidgeClassifierCV(alphas=np.logspace(-3, 3, 10))
    classifier.fit(X_train_transform, y_train)
    X_val_transform = minirocket.transform(X_val)
    y_val_pred = classifier.predict(X_val_transform)
    return Metric(y_val, y_val_pred, binary=True)
```

```python
cross_val(context, data, baseline_minirocket)
```
